```python
#import all the libraries necessary to combine two images into one

from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
from PIL import Image
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
# Load the image
image_path = '/content/unicorn.webp'  # Replace with the actual path to your image
image = cv2.imread(image_path)
image
```

## RESIZING

```python
# Resizing
resized_image = cv2.resize(image, (224, 224))
cv2_imshow(resized_image)
```

## NORMALIZATION

```python
# Normalization
normalized_image = resized_image / 255.0
cv2_imshow(normalized_image)
```

##DATA AUGMENTATION

```python
# Data Augmentation
data_generator = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
augmented_images = []
for i in range(5):  # Generate 5 augmented images
    augmented_image = data_generator.random_transform(resized_image)
    augmented_images.append(augmented_image)
for img in augmented_images:
    cv2_imshow(img)
```

##CROPPING

```python
# Cropping
cropped_image = resized_image[50:170, 50:170]
cv2_imshow(cropped_image)
```

##GRAYSCALE

```python
# Grayscale
gray_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

##HISTOGRAM EQUALIZATION

```python
# Histogram Equalization
equalized_image = cv2.equalizeHist(gray_image)
cv2_imshow(equalized_image)
```

##GAUSSIAN BLURRING

```python
# Gaussian Blurring
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
cv2_imshow(blurred_image)
```

##IMAGE ROTATION

```python
# Image Rotation
rows, cols = image.shape[:2]
rotation_matrix = cv2.getRotationMatrix2D((cols / 2, rows / 2), 45, 1)
rotated_image = cv2.warpAffine(image, rotation_matrix, (cols, rows))
cv2_imshow(rotated_image)
```

##NOISE REDUCTION

```python
# Noise Reduction
noisy_image = cv2.imread(image_path)
# Apply noise reduction techniques, for example, using Gaussian Blur
denoised_image = cv2.GaussianBlur(noisy_image, (5, 5), 0)
cv2_imshow(denoised_image)
```

##EDGE DETECTION

```python
# Edge Detection
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray_image, 100, 200)
cv2_imshow(edges)
```

##COLOR SPACE CONVERSION

```python
# Color Space Conversion
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
cv2_imshow(hsv_image)
```

##CONTRAST ADJUSTMENT

```python
# Contrast Adjustment
alpha = 1.5  # Contrast control (1.0-3.0)
beta = 0    # Brightness control (0-100)
adjusted_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)
cv2_imshow(adjusted_image)
```
